```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data  = pd.read_csv(r"https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv")
#night price distribution of Airbnb appartments 
#count vs price 

# data %>%
#   filter( price<300 ) %>%
#   ggplot( aes(x=price)) +
#     stat_bin(breaks=seq(0,300,10), fill="#69b3a2", color="#e9ecef", alpha=0.9) +
#     ggtitle("Night price distribution of Airbnb appartements") +
#     theme_ipsum() +
#     theme(
#       plot.title = element_text(size=12)
#     )

#plot histogram in python , price < 300
data = data[data['price']<300]
plt.hist(data['price'], bins=30, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
plt.title('Night price distribution of Airbnb appartements')
plt.show()

```

```python
# data %>%
#   filter( price<300 ) %>%
#   ggplot( aes(x=price)) +
#     geom_density(fill="#69b3a2", color="#e9ecef", alpha=0.7, bw=10) +
#     ggtitle("Bandwidth: 10") +
#     theme_ipsum()
#density plot 
data = data[data['price']<300]
sns.distplot(data['price'], hist=False, color='#69b3a2', kde_kws={"shade": True})
plt.title('Bandwidth: 10')
```

```python
# data %>%
#   filter( price<300 ) %>%
#   ggplot( aes(x=price)) +
#     geom_density(fill="#69b3a2", color="#e9ecef", alpha=0.7, bw=2) +
#     ggtitle("Bandwidth: 2") +
#     theme_ipsum()

#bandwidth 2
data = data[data['price']<300]
sns.displot(data['price'], kind = 'kde', bw_adjust=0.2, color='#69b3a2', fill=True, label='price')
plt.title('Bandwidth: 10')
plt.show()
```
