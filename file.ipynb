```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data  = pd.read_csv(r"https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/1_OneNum.csv")
#night price distribution of Airbnb appartments 
#count vs price 

data = data[data['price']<300]
plt.hist(data['price'], bins=30, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
plt.title('Night price distribution of Airbnb appartements')
plt.show()

```

```python


data = data[data['price']<300]
plt.hist(data['price'], bins=100, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
plt.title('Night price distribution of Airbnb appartements')
plt.show()
```

APARTMENT PRICE VS GROUND LIVING AREA.

```python
data2 = pd.read_csv(r"https://raw.githubusercontent.com/holtzy/data_to_viz/master/Example_dataset/2_TwoNum.csv")

#plot 2 histograms
plt.hist(data2['GrLivArea'], bins=50, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
plt.title('Ground living area distribution')
plt.xlabel('area')
plt.show()

plt.hist(data2['SalePrice']/1000, bins=50, color='#69b3a2', edgecolor='#e9ecef', alpha=0.9)
plt.title('Sale price distribution')
plt.xlabel('Sale price (k$)')
plt.show()
```

```python
#scatterplot 

plt.scatter(data2['GrLivArea'], data2['SalePrice']/1000, color='#69b3a2', alpha=0.8)
plt.title('Ground living area partially explains sale price of apartments')
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.show()
```

```python
#improving the scatter plot 

scatter = plt.scatter(data2['GrLivArea'], data2['SalePrice']/1000, color='#69b3a2', alpha=0.8)
plt.title('Ground living area partially explains sale price of apartments')
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.show()
#p1

#linear regression
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(data2['GrLivArea'], data2['SalePrice'])
line = slope*data2['GrLivArea']+intercept
plt.plot(data2['GrLivArea'], line, 'black', label='fitted line')
plt.legend()
plt.show()





```

```python
#loess with scatter plot 
scatter = plt.scatter(data2['GrLivArea'], data2['SalePrice']/1000, color='#69b3a2', alpha=0.8)
plt.title('Ground living area partially explains sale price of apartments')
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.show()
#loess 
from statsmodels.nonparametric.smoothers_lowess import lowess
lowess = lowess(data2['SalePrice']/1000, data2['GrLivArea'], frac=0.3)
plt.plot(lowess[:,0], lowess[:,1], color='black', label='fitted line')
plt.legend()
plt.show()
```

```python
#reduce dot size 
scatter = plt.scatter(data2['GrLivArea'], data2['SalePrice']/1000, color='#69b3a2', alpha=0.8, s=1)
plt.title('Ground living area partially explains sale price of apartments')
plt.xlabel('Ground living area')
plt.ylabel('Sale price (k$)')
plt.show()
```

```python
sns.kdeplot(x=data2["GrLivArea"], y=data2["SalePrice"],color="#69b3a2")
plt.show()
```

```python
#add area 
sns.kdeplot(x=data2["GrLivArea"], y=data2["SalePrice"],color="#69b3a2", shade=True)
plt.show()
```

```python
#hexbin 
plt.hexbin(data2['GrLivArea'], data2['SalePrice']/1000, gridsize=30, cmap='viridis', mincnt=1)
plt.show()
```

```python
#2d histogram
plt.hist2d(data2['GrLivArea'], data2['SalePrice']/1000, bins=30, cmap='viridis', cmin=1)
plt.show()
```
